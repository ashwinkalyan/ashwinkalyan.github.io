```python
import scipy.optimize as opt
import numpy as np
from tqdm import tqdm

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
obj = []
num_trials = 50
for n in tqdm(range(4,50)):
    temp = []
    for _ in range(num_trials):
        W = np.random.uniform(0,1,(n,n))
        r,c = opt.linear_sum_assignment(W)
        cost = W[r,c].sum()
        temp.append(cost)
    obj.append(sum(temp)/num_trials)
```

```python
plt.plot(range(4,50), obj)
plt.grid()
plt.xlabel('|U|=|V|=n')
plt.ylabel('LP cost averaged over 20 runs')
```
